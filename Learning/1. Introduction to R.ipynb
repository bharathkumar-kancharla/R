# Help in R:

    1) Start R software and click the help button in the toolbar of the R Gui (Graphic user interface) window.
    2) Search for help in Google www.google.com
    3) If you need help with a function, then type question mark followed by the name of the function. For example, ?read.table to get help for function read.table.
    4) Sometimes, you want to search by the subject on which we want help (e.g. data input). In such a case, type help.search("data input")
    5) 'help()' for on-line help, or 'help.start()’ for an HTML browser interface to help.
    6) The find function tells us what package something is in.
        Ex: find("lowess") returns  [1] "package:stats"
    7) The apropos returns a character vector giving the names of all objects in the search list that match your enquiry.
        Ex: apropos("lm") returns list of commands which contain lm
    8) To see worked examples – example
        Ex: example(lm) --> we see the printed and graphical output produced by the lm function.
    9) demo(persp) --> This can be useful for seeing the type of things that R can do. 
        Ex: [persp is a command for 3d surface plots]
    10) 'q()' to quit R.

# Cleaning up the Windows:

- We assign names to variables when analyzing any data. It is good practice to remove the variable names given to any data frame at the end each session in R.

- This way, variables with same names but different properties will not get in each others way in subsequent work.

- `rm()` command removes variable names

**Ex:** rm(x,y,z) removes the variables x, y and z.

- `attach(bateria)` --> attaches objects to the search path
- `detach()` --> command detaches objects from the Search Path, It removes it from the `search()` path of available R objects.

- `rm(list=ls())`  --> To get rid of everything, including data frames
